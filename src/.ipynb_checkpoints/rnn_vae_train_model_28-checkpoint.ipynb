```python
X_train = np.load(root+"/X_train.npy")
X_val = np.load(root+"/X_val.npy")
X_test = np.load(root+"/X_test.npy")
```

```python
X_train = ae.encode_series(rnn_data_loader.X_train)
X_val = ae.encode_series(rnn_data_loader.X_val)
X_test = ae.encode_series(rnn_data_loader.X_test)
```

```python
X_train = ae.encode_series(X_train)
X_val = ae.encode_series(X_val)
X_test = ae.encode_series(X_test)
```

```python
seq = Seq2Seq()
seq.build_model(input_length=20, input_dim=64, latent_dim=(256,), output_length=30, output_dim=64)
```

```python
seq.set_weights(root+"/../models/seq2seq_vae_v1.h5")
```

```python
# TRAIN LOOP
for epoch in range(10000):
    print('EPOCH %d' % (epoch))
    seq.train(X_train=X_train[:, :20, :], 
             Y_train=X_train[:, 20:50, :], 
             X_val=X_val[:, :20, :], 
             Y_val=X_val[:, 20:50, :], 
             epochs=1, 
             batch_size=32
             )
    if epoch % 500 == 0:
        i = random.randint(0, X_train.shape[0]-1)
        X, Y = [], []
        x = X_train[i, :20, :]
        X.append(x)
        X = np.array(X)
        y = X_train[i, 20:50, :]
        Y.append(y)
        Y = np.array(Y)
        Y_hat = seq.model.predict(X)
        result = np.concatenate((X, Y_hat), axis=1)
        Y_hat_decoded = ae.decode_series(result)
        show_images_as_video(Y_hat_decoded[0])
    seq.save_weights(root+"/../models/seq2seq_vae_v1.h5")
```

```python
def produce_videos(train=True, val=False, test=False, i=None):
    if i == None:
        if train:
            i = random.randint(0, X_train.shape[0]-1)
        elif val:
            i = random.randint(0, X_val.shape[0]-1)
        elif test:
            i = random.randint(0, X_test.shape[0]-1)
    X, Y = [], []
    x, y = None, None
    if train:
        x = X_train[i, :20, :]
        y = X_train[i, 20:50, :]
    elif val:
        x = X_val[i, :20, :]
        y = X_val[i, 20:50, :]
    elif test:
        x = X_test[i, :20, :]
        y = X_test[i, 20:50, :]
    X.append(x)
    X = np.array(X)
    Y.append(y)
    Y = np.array(Y)
    Y_hat = seq.model.predict(X)
    result = np.concatenate((X, Y_hat), axis=1)
    Y_hat_decoded = ae.decode_series(result)
    result = np.concatenate((X, Y), axis=1)
    Y_decoded = ae.decode_series(result)
    return Y_decoded[0], Y_hat_decoded[0], i
```

```python
def produce_long_videos(train=True, val=False, test=False, i=None, n=1):
    if i == None:
        if train:
            i = random.randint(0, X_train.shape[0]-1)
        elif val:
            i = random.randint(0, X_val.shape[0]-1)
        elif test:
            i = random.randint(0, X_test.shape[0]-1)
    X, Y = [], []
    x, y = None, None
    if train:
        x = X_train[i, :20, :]
        y = X_train[i, 20:50, :]
    elif val:
        x = X_val[i, :20, :]
        y = X_val[i, 20:50, :]
    elif test:
        x = X_test[i, :20, :]
        y = X_test[i, 20:50, :]
    X.append(x)
    X = np.array(X)
    Y_hat = seq.model.predict(X)
    result = np.concatenate((X, Y_hat), axis=1)
    for j in range(n):
        X_in = result[:, -20:, :]
        Y_hat_out = seq.model.predict(X_in)
        result = np.concatenate((result, Y_hat_out[:, :, :]), axis=1)    
    print(result.shape)
    Y_hat_decoded = ae.decode_series(result)
    if train:
        y = X_train[i, :result.shape[1], :]
    elif val:
        y = X_val[i, :result.shape[1], :]
    elif test:
        y = X_test[i, :result.shape[1], :]
    Y.append(y)
    Y = np.array(Y)    
    Y_decoded = ae.decode_series(Y)
    return Y_decoded[0], Y_hat_decoded[0], i
```

```python
def get_params(i):
    i = random.randint(0, X_train.shape[0]-1)
    X, Y = [], []
    x = X_train[i, :20, :]
    X.append(x)
    X = np.array(X)
    y = X_train[i, 20:50, :]
    Y.append(y)
    Y = np.array(Y)
    mu, log_var = seq.encoder_mu_log_var.predict(X)
    Y_hat = seq.model.predict(X)
    result = np.concatenate((X, Y_hat), axis=1)
    Y_hat_decoded = ae.decode_series(result)
    # show_images_as_video(Y_hat_decoded[0])
    return mu, log_var
```

```python
balls = {
    1 : [78, 12, 20],
    2: [5, 16, 66],
    3: [36, 22],
    4: [46],
    5: [8]
}

params = {
    1 : [],
    2: [],
    3: [],
    4: [],
    5: []
}

dist = {}

for key, value in balls.items():
    for i in value:
        params[key].append(get_params(i))
```

```python
mu, log_var = params[1][0]
for key, value in params.items():
    n_mu, n_log_var = params[key][0]
    dist[key] = (np.linalg.norm(mu-n_mu), np.linalg.norm(log_var-n_log_var))
```

```python
'''
Conclusions:

The variational rnn is generative. Therefore, it is able to produce new videos from no input by sampling the latent space. This is expected and the intent of the a seq2seq vae.

However, I hypothesized that a vae will outperform the deterministic rnn in producing frames past what it was trained for.
This is because there is only one correct future in a deterministic rnn thereby making it much harder to extrapolate to the correct future after the 30 timestep prediciton during training.

The vae rnn is able to generate more than 30 frames by taking the last twenty it produced as input to the model again.
What is noticed is the vae rnn has inferior performance on balls fewer than or equal to two compared to the deterministic rnn. 
This might be because of the distribution of the training data (more videos containing more than two balls).

When the vae generates past 30 timesteps, they are not aligned with the past frames. In other words, it looks like a new video altogether.
While the determinstic rnn could not even produce a new accurate video, the vae can although it does not follow the motion of the last 20 timesteps.
However, it is not entirely new either. The frames past 30 timesteps, although they look like a new video, always tend to have the same number of balls 
(+-1) in the new frames that it was not trained on. This leads me to believe that the model learns to recognize and associate videos with similar number of balls.
'''
```
