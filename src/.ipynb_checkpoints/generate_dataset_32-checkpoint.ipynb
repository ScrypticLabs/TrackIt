```python
import os
import numpy as np
from data_loader import DataLoader
from utils.visualize import show_images_in_grid
```

```python
root = os.getcwd()
data_dir = root+"/../datasets/data/32/"
data_loaders = {}
data_loaders[0] = DataLoader()
```

```python
for i in range(1, 6):
    data_loaders[i] = DataLoader(features_dir=data_dir+str(i)+".npy", labels_dir=data_dir+str(i)+".npy")
    data_loaders[i].split_train_val_test(train=0.8, val=0.1, test=0.1)
    if i == 1:
        data_loaders[0].X_train = np.copy(data_loaders[i].X_train[:, :100, :, :, :])
        data_loaders[0].Y_train = np.copy(data_loaders[i].Y_train[:, :100, :, :, :])
        data_loaders[0].X_val = np.copy(data_loaders[i].X_val[:, :100, :, :, :])
        data_loaders[0].Y_val = np.copy(data_loaders[i].Y_val[:, :100, :, :, :])
        data_loaders[0].X_test = np.copy(data_loaders[i].X_test[:, :100, :, :, :])
        data_loaders[0].Y_test = np.copy(data_loaders[i].Y_test[:, :100, :, :, :])
    else:
        data_loaders[0].X_train = np.concatenate((data_loaders[0].X_train, data_loaders[i].X_train[:, :100, :, :, :]))
        data_loaders[0].Y_train = np.concatenate((data_loaders[0].Y_train, data_loaders[i].Y_train[:, :100, :, :, :]))
        data_loaders[0].X_val = np.concatenate((data_loaders[0].X_val, data_loaders[i].X_val[:, :100, :, :, :]))
        data_loaders[0].Y_val = np.concatenate((data_loaders[0].Y_val, data_loaders[i].Y_val[:, :100, :, :, :]))
        data_loaders[0].X_test = np.concatenate((data_loaders[0].X_test, data_loaders[i].X_test[:, :100, :, :, :]))
        data_loaders[0].Y_test = np.concatenate((data_loaders[0].Y_test, data_loaders[i].Y_test[:, :100, :, :, :]))
```

```python
np.save(root+"/../datasets/train/32/X_train", data_loaders[0].X_train)
np.save(root+"/../datasets/train/32/Y_train", data_loaders[0].Y_train)
np.save(root+"/../datasets/val/32/X_val", data_loaders[0].X_val)
np.save(root+"/../datasets/val/32/Y_val", data_loaders[0].Y_val)
np.save(root+"/../datasets/test/32/X_test", data_loaders[0].X_test)
np.save(root+"/../datasets/test/32/Y_test", data_loaders[0].Y_test)
```

```python
data_loaders[0].X_train = np.load(root+"/../datasets/train/32/X_train.npy")
data_loaders[0].Y_train = np.load(root+"/../datasets/train/32/Y_train.npy")
data_loaders[0].X_val = np.load(root+"/../datasets/validation/32/X_val.npy")
data_loaders[0].Y_val = np.load(root+"/../datasets/validation/32/Y_val.npy")
data_loaders[0].X_test = np.load(root+"/../datasets/test/32/X_test.npy")
data_loaders[0].Y_test = np.load(root+"/../datasets/test/32/Y_test.npy")
```
