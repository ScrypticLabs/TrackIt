```python
import os
import numpy as np
from data_loader import DataLoader
from utils.visualize import show_images_in_grid, show_reconstructions
from cnn import Autoencoder
import random
```

```python
root = os.getcwd()
cnn_train_dir = root+"/../datasets/training_set/data/cnn.npy"
cnn_test_dir = root+"/../datasets/test_set/data/cnn.npy"
```

```python
# WE WON'T USE THE TEST DATA AT ALL AND INSTEAD SPLIT THE TRAIN DATA INTO TRAIN, TEST, AND VALIDATION DATASETS WITH THE DATA LOADER
```

```python
cnn_data_loader = DataLoader(features_dir=cnn_train_dir, labels_dir=cnn_train_dir)
cnn_data_loader.split_train_val_test(train=0.7, val=0.2, test=0.1)
# NORMALIZE ALL PIXEL VALUES FROM 0-255 TO O-1
cnn_data_loader.preprocess(rescale=True)
```

```python
show_images_in_grid(images=cnn_data_loader.X_train, columns=10, rows=5, img_size=(50,50), plot_size=(20,10), rescale=True)
```

```python
ae = Autoencoder()
ae.build_model(input_dim=(28, 28, 3), latent_dim=(64,))
```

```python
ae.encoder.summary()
```

```python
ae.decoder.summary()
```

```python
ae.set_weights(root+"/../models/autoencoder.h5")
```

```python
Y_train_hat = ae.autoencoder.predict(cnn_data_loader.X_train)
Y_val_hat = ae.autoencoder.predict(cnn_data_loader.X_val)
Y_test_hat = ae.autoencoder.predict(cnn_data_loader.X_test)
```

```python
# TRAIN
```

```python
show_reconstructions(cnn_data_loader.X_train, Y_train_hat, 10, (50,50), (20,4), rescale=True)
```

```python
# VALIDATION
```

```python
show_reconstructions(cnn_data_loader.X_val, Y_val_hat, 10, (50,50), (20,4), rescale=True)
```

```python
# TEST
```

```python
show_reconstructions(cnn_data_loader.X_test, Y_test_hat, 10, (50,50), (20,4), rescale=True)
```
